```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt # importing matplotlib
import numpy as np # importing numpy
```

```python
df=pd.read_csv("Data/df_train_tratado.csv", sep=';')
```

```python
df.head()
```

```python
cols = ['classe','sexo','local_embarque','cabine_lvl','tem_cabine']

for col in cols:
    fig, ax = plt.subplots(figsize=(15, 7))

    #remover grid
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    sns.barplot(x=col, y='sobreviveu', 
                data=df, 
                estimator=sum,              
                ci=None,
                color='c').set_title(f'Qtd. de Sobreviventes por {col}')

    plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14, 7))

sns.set(style=None)
sns.histplot(data=df, x='idade', hue='sobreviveu', stat='count', edgecolor=None)

#label eixo x
ax.set_ylabel('Frequncia')
ax.set_xlabel('Idade')

#remover grid
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.show()
```
