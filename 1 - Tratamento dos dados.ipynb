```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def verifica_nulo(df):
    """
    Mostra as colunas que tem null e que não tem e suas porcentagens.
    """
    for column in df.columns:
        null = df[column].isnull().sum()
        p = (null/(df.shape[0]))*100
        if df[column].isnull().sum()==0:
            continue
        else:
            print(f'{column}: {p.round(2)}% or {null} nulls',)
    print( )
    
    print('The columns with the most % nulls are:')
    print(((df.isnull().sum()/df.shape[0])*100).sort_values(ascending = False))
```

```python
def verifica_linhas_duplicadas(df):
    """
    informa se existem linhas duplicadas
    """
    # Linhas duplicadas
    if df.shape[0] != df.drop_duplicates().shape[0]:
        n_linhas_duplicadas = df.shape[0] - df.drop_duplicates().shape[0]
        print(f'Existem {n_linhas_duplicadas} linhas duplicadas')
    else:
        print('Não há linhas duplicadas no conjunto de dados.')
```

```python
def verifica_colunas_duplicadas(df):
    """
    informa se existem colunas duplicadas
    """
    # Colunas duplicadas
    colunas_duplicadas = []
    for col1 in df.columns.tolist():
        for col2 in df.columns.tolist():
            if col1 != col2:
                if (df[col1] == df[col2]).all():
                    colunas_duplicadas.append([col1, col2])

    if len(colunas_duplicadas) > 0:
        print('Colunas duplicadas:', colunas_duplicadas)
    else:
        print('Não há colunas duplicadas no conjunto de dados.')
```

```python
## Verificar as linhas com o local embarque nulo
df_train_raw.loc[df_train_raw['local_embarque'].isnull()]
```

```python
#existe 127 passegieros da classe 1 que embarcaram em S(Southampton)
df_train_raw.loc[df_train_raw['classe']==1].groupby(by="local_embarque").count()
```

```python
#Preeencher os o local embarque nulo com S
df_train_raw['local_embarque'] = df_train_raw['local_embarque'].fillna('S')
```

```python
classe_group_count = {}

for classe in df_train_raw['classe'].unique():
    aux = df_train_raw.loc[df_train_raw['classe'] == classe]
    count_cabine = aux['cabine'].count()
    classe_group_count[f'Classe {classe} '] = f'{count_cabine} Cabines Preenchidas'
```

```python
classe_group_count
```

```python
#Criando a coluna com o tipo cabine
df_train_raw['cabine_lvl'] = df_train_raw['cabine'].str.extract('([A-Za-z]+)', expand = True)
```

```python
df_train_raw
```

```python
df_train_raw['cabine_lvl'].unique()
```

```python
#Preeencher o nivel da cabine com Z para os NAN
df_train_raw['cabine_lvl'] = df_train_raw['cabine_lvl'].fillna('Z')
```

```python
#Criar coluna nova tem_cabine
df_train_raw['tem_cabine'] = np.where(df_train_raw['cabine'].isnull(), 0, 1)
```

```python
#apagar coluna cabine
df_train_raw.drop('cabine', axis=1, inplace=True)
```

```python
#verificando a idade
df_train_raw['idade'].unique()
```

```python
age_group_map = {}

for group in df_train_raw['sexo'].unique():
    aux = df_train_raw.loc[df_train_raw['sexo'] == group]
    median_age = aux['idade'].median()
    age_group_map[group] = median_age
```

```python
age_group_map
```

```python
#preenchendo a idade com a média por sexo
cond_age_null = df_train_raw['idade'].isnull()
df_train_raw.loc[cond_age_null, 'idade'] = df_train_raw.loc[cond_age_null, 'sexo'].map(age_group_map)
```

```python
#dropando a coluna Ticket
df_train_raw.drop('Ticket', axis=1, inplace=True)
```
